## Create Employees DataFrame

### Subtask:
Create a pandas DataFrame named `employees_df` with columns for 'Employee_ID', 'Employee_Name', 'Salary', and 'Department', using the provided employee data. Ensure data is correctly structured and free of syntax errors.

```python
import pandas as pd

# Create a sample DataFrame for employees
data = {
    'Employee_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Employee_Name': ['Abhishek', 'Shubham', 'Priya', 'Rohit', 'Neha', 'Aman', 'Ravi', 'Sneha', 'Kiran', 'Tanuja'],
    'Salary': [62000, 58000, 67000, 90000, 55000, 60000, 75000, 70000, 65000, 80000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT']
}
employees_df = pd.DataFrame(data)

print("DataFrame 'employees_df' created successfully:")
print(employees_df.head())
```

## Calculate Average Salary Per Department

### Subtask:
Group the `employees_df` by 'Department' and calculate the average 'Salary' for each department.


**Reasoning**:
To calculate the average salary per department, I will group the `employees_df` by the 'Department' column and then compute the mean of the 'Salary' for each group.



```python
average_salary_by_department = employees_df.groupby('Department')['Salary'].mean()
```

## Identify Department with Highest Average Salary

### Subtask:
Determine which department has the highest average salary from the previously calculated departmental averages.


**Reasoning**:
To find the department with the highest average salary, I will use the `.idxmax()` method on the `average_salary_by_department` Series, store the result, and then print it as instructed.



```python
highest_paying_department = average_salary_by_department.idxmax()
print(f"Department with the highest average salary: {highest_paying_department}")
```

## Filter Employees by Department

Filter the original `employees_df` to include only those employees who belong to the department identified as having the highest average salary.


**Reasoning**:
To achieve the subtask, I will filter the `employees_df` DataFrame using boolean indexing to select employees whose 'Department' matches the `highest_paying_department` and store the result in `employees_in_highest_paying_dept`, then print the head of this new DataFrame.



```python
employees_in_highest_paying_dept = employees_df[employees_df['Department'] == highest_paying_department]
print(f"Employees in the {highest_paying_department} department:")
print(employees_in_highest_paying_dept.head())
```
